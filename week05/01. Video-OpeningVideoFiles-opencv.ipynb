# Display video from Notebook

```python
import cv2
from IPython.display import display, Image, clear_output
import numpy as np

# Open the video file
cap = cv2.VideoCapture('./videos/butterbear.mp4')


if not cap.isOpened():
    print("Cannot open video file")
    exit()

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        print("Stream stopped.")
        break

    # Convert the frame to JPEG format for display in Jupyter
    _, buffer = cv2.imencode('.jpg', frame)
    img_bytes = buffer.tobytes()

    # Display the frame in Jupyter Notebook
    display(Image(data=img_bytes))
    clear_output(wait=True)  # Clear previous frame for smoother playback

  
    
cap.release()
print("Video stream ended.")
```

```python
import cv2
from IPython.display import display, Image, clear_output
import numpy as np

# Open both video files
cap1 = cv2.VideoCapture('./videos/butterbear.mp4')
cap2 = cv2.VideoCapture('./videos/drummer.mp4')

if not cap1.isOpened() or not cap2.isOpened():
    print("Cannot open one or both video files")
    exit()

# Get frame dimensions (use first video's dimensions as reference)
width = int(cap1.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap1.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Resize target for consistent display
target_width = 320
target_height = 240

while True:
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()
    
    # If both videos ended, stop
    if not ret1 and not ret2:
        print("Both streams ended.")
        break
    
    # Create black frame if video ended
    if ret1:
        frame1 = cv2.resize(frame1, (target_width, target_height))
    else:
        frame1 = np.zeros((target_height, target_width, 3), dtype=np.uint8)
    
    if ret2:
        frame2 = cv2.resize(frame2, (target_width, target_height))
    else:
        frame2 = np.zeros((target_height, target_width, 3), dtype=np.uint8)
    
    # Combine frames side by side
    combined = np.hstack((frame1, frame2))
    
    # Convert to JPEG for display
    _, buffer = cv2.imencode('.jpg', combined)
    img_bytes = buffer.tobytes()
    
    display(Image(data=img_bytes))
    clear_output(wait=True)

cap1.release()
cap2.release()
print("Video stream ended.")
```
