```python
from pytubefix import YouTube
import cv2

# ==================== ตั้งค่าตรงนี้ ====================
video_url = "https://youtu.be/EsHQB9gT96k?si=bdi3dDrVnYVIgVYi"
output_filename = "my_video.mp4"  # ตั้งชื่อไฟล์ที่ต้องการ
start_time = 75   # เริ่มที่วินาที
end_time = 120    # จบที่วินาที
# ======================================================

# Get stream URL
yt = YouTube(video_url)
print(f"Title: {yt.title}")

video_stream = yt.streams.filter(file_extension='mp4', progressive=True).get_highest_resolution()

if not video_stream:
    print("No compatible video stream found.")
    exit()

stream_url = video_stream.url

# Open video stream
cap = cv2.VideoCapture(stream_url)

if not cap.isOpened():
    print("Cannot open video stream")
    exit()

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

print(f"FPS: {fps}, Resolution: {width}x{height}")
print(f"Saving to: {output_filename}")

# Setup video writer
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_filename, fourcc, fps, (width, height))

# Set start position
cap.set(cv2.CAP_PROP_POS_MSEC, start_time * 1000)

frame_count = 0
while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        break
    
    current_time = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000
    
    if current_time >= end_time:
        break
    
    out.write(frame)
    frame_count += 1
    
    if frame_count % 30 == 0:
        print(f"\rProcessing: {current_time:.1f}s / {end_time}s", end="")

cap.release()
out.release()
print(f"\nSaved {frame_count} frames to {output_filename}")
```

# Save Video

```python
import cv2
from IPython.display import display, Image, clear_output
import numpy as np

# Open the video file
cap = cv2.VideoCapture(output_filename)


if not cap.isOpened():
    print("Cannot open video file")
    exit()

while cap.isOpened():
    ret, frame = cap.read()
    
    if not ret:
        print("Stream stopped.")
        break

    # Convert the frame to JPEG format for display in Jupyter
    _, buffer = cv2.imencode('.jpg', frame)
    img_bytes = buffer.tobytes()

    # Display the frame in Jupyter Notebook
    display(Image(data=img_bytes))
    clear_output(wait=True)  # Clear previous frame for smoother playback

  
    
cap.release()
print("Video stream ended.")
```
