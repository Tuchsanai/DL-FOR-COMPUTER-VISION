```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

path  ="./Data"
files = os.listdir(path)
image_files = [f for f in files if f.endswith('.jpg')]


print(image_files)
```

```python


fig, ax = plt.subplots(2, 2, figsize=(15, 10))  # Create a 2x2 grid of subplots

fig.subplots_adjust(wspace=0.1, hspace=0.01)  # Set wspace and hspace to create a tiny space


for i in range(2):
    for j in range(2):
        img = cv2.imread(os.path.join(path, image_files[i * 2 + j]))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        ax[i, j].imshow(img)  # Display the image in the corresponding subplot
        ax[i, j].axis('off')  # Turn off the axis for a cleaner look

plt.show()  # Show the grid of images
```
