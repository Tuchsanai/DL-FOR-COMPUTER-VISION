```python
import cv2  
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('banana.jpg')
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

```

```python

# dispaly of two image using matplotlib using subplot
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.subplot(1,2,2)
plt.title('HSV Image')
plt.imshow(cv2.cvtColor(hsv_image, cv2.COLOR_BGR2RGB))
```

Determining Color Range: The HSV value ranges for yellow (ripe banana) and green (unripe banana) colors are determined.

```python
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([30, 255, 255])
lower_green = np.array([35, 100, 100])
upper_green = np.array([85, 255, 255])
```

```python
# Mask for yellow banana
mask_yellow = cv2.inRange(hsv_image, lower_yellow, upper_yellow)


# Mask for green banana
mask_green = cv2.inRange(hsv_image, lower_green, upper_green)


# display of two mask using matplotlib using subplot
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.title('Mask for Yellow Banana')
plt.imshow(mask_yellow)
plt.subplot(1,2,2)
plt.title('Mask for Green Banana')
plt.imshow(mask_green)
```

```python
# Count the number of pixels for each mask
yellow_count = cv2.countNonZero(mask_yellow)
green_count = cv2.countNonZero(mask_green)

total_pixels = hsv_image.shape[0] * hsv_image.shape[1]

# Determine ripeness
if yellow_count > green_count and yellow_count > 0.1 * total_pixels:
    ripeness = 'Ripe'
elif green_count > yellow_count and green_count > 0.1 * total_pixels:
    ripeness = 'Unripe'
else:
    ripeness = 'Undetectable'


print(f'The banana is {ripeness} with {yellow_count} yellow pixels and {green_count} green pixels')
```
