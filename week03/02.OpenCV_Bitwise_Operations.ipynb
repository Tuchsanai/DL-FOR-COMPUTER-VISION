# Bitwise Operations in OpenCV

## Explanation:

Bitwise operations work on the binary representation of pixel values.  
Each pixel in an image has an intensity value (0-255 for 8-bit grayscale images). These values are represented in binary.  

OpenCV provides the following bitwise operations:
1. **Bitwise AND** (`cv2.bitwise_and`)
2. **Bitwise OR** (`cv2.bitwise_or`)
3. **Bitwise XOR** (`cv2.bitwise_xor`)
4. **Bitwise NOT** (`cv2.bitwise_not`)

#


```python

# Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Helper function to display images
def display_images(images, titles):
    plt.figure(figsize=(12, 8))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.show()
```


### Step 1: Create Sample Images for Bitwise Operations
We will create two black images with white rectangles to demonstrate the bitwise operations.


```python

# Create two black images
img1 = np.zeros((300, 300), dtype=np.uint8)
img2 = np.zeros((300, 300), dtype=np.uint8)

# Draw white rectangles on the images
cv2.rectangle(img1, (50, 50), (250, 250), 255, -1)  # White square on img1
cv2.rectangle(img2, (150, 150), (300, 300), 255, -1)  # White square on img2

# Display images
display_images([img1, img2], ["Image 1", "Image 2"])
```


### Step 2: Perform Bitwise Operations
#### Bitwise AND
Only the overlapping white regions in both images are retained.


```python

# Bitwise AND
and_result = cv2.bitwise_and(img1, img2)
display_images([img1, img2, and_result], ["Image 1", "Image 2", "Bitwise AND"])
```


#### Bitwise OR
All white regions from both images are combined.


```python

# Bitwise OR
or_result = cv2.bitwise_or(img1, img2)
display_images([img1, img2, or_result], ["Image 1", "Image 2", "Bitwise OR"])
```


#### Bitwise XOR
White regions that are unique to each image are retained.


```python

# Bitwise XOR
xor_result = cv2.bitwise_xor(img1, img2)
display_images([img1, img2, xor_result], ["Image 1", "Image 2", "Bitwise XOR"])
```


#### Bitwise NOT
White becomes black, and black becomes white.


```python

# Bitwise NOT
not_result = cv2.bitwise_not(img1)
display_images([img1, not_result], ["Image 1", "Bitwise NOT"])
```


### Step 3: Practical Example - Masking an Image
We will load an image, create a circular mask, and apply it using `cv2.bitwise_and`.


```python

# Load a sample image
image = cv2.imread(cv2.samples.findFile('.././images/Taj_Mahal.jpg'), cv2.IMREAD_GRAYSCALE)

# Create a circular mask
mask = np.zeros_like(image, dtype=np.uint8)
cv2.circle(mask, (mask.shape[1] // 2, mask.shape[0] // 2), 150, 255, -1)

# Apply the mask using Bitwise AND
masked_image = cv2.bitwise_and(image, image, mask=mask)

# Display results
display_images([image, mask, masked_image], ["Original Image", "Mask", "Masked Image"])
```
