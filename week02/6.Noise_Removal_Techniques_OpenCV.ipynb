# Noise Removal Techniques using OpenCV


In this notebook, we will explore four noise removal techniques using OpenCV:
- **Average Blur**
- **Gaussian Blur**
- **Median Blur**
- **Bilateral Filter**

Each technique will be described mathematically and demonstrated with code.
    

## Noise Removal Techniques with OpenCV

| **Filter Name**      | **OpenCV Code**                          | **Description**                                                                                  |
|-----------------------|------------------------------------------|--------------------------------------------------------------------------------------------------|
| **Average Blur**      | `cv2.blur(image, (k, k))`               | Averages the pixel values in a \( k \times k \) kernel. Reduces noise but blurs edges.           |
| **Gaussian Blur**     | `cv2.GaussianBlur(image, (k, k), sigmaX)` | Applies a Gaussian kernel. Reduces noise while retaining more details compared to Average Blur. |
| **Median Blur**       | `cv2.medianBlur(image, k)`              | Replaces each pixel with the median value in its \( k \times k \) kernel. Effective for salt-and-pepper noise. |
| **Bilateral Filter**  | `cv2.bilateralFilter(image, d, sigmaColor, sigmaSpace)` | Combines spatial and intensity weights to reduce noise while preserving edges.                  |

### Parameters
- **`k`**: Kernel size (odd numbers, e.g., 3, 5, 7).
- **`sigmaX`**: Standard deviation for Gaussian Blur.
- **`d`**: Diameter of the pixel neighborhood for Bilateral Filter.
- **`sigmaColor`**: Intensity-based filtering strength for Bilateral Filter.
- **`sigmaSpace`**: Spatial filtering strength for Bilateral Filter.


```python

# Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the noisy image
image_path = ".././images/test_noise.jpg"  # Replace with your image path
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image
plt.imshow(image_rgb)
plt.title('Original Image with Noise')
plt.axis('off')
plt.show()
    
```

## Average Blur

```python

# Apply Average Blur
average_blur = cv2.blur(image_rgb, (5, 5))

# Display the result
plt.imshow(average_blur)
plt.title('Average Blur')
plt.axis('off')
plt.show()
    
```

## Gaussian Blur

```python

# Apply Gaussian Blur
gaussian_blur = cv2.GaussianBlur(image_rgb, (5, 5), 0)

# Display the result
plt.imshow(gaussian_blur)
plt.title('Gaussian Blur')
plt.axis('off')
plt.show()
    
```

```python

# Apply Median Blur
median_blur = cv2.medianBlur(image_rgb, 5)

# Display the result
plt.imshow(median_blur)
plt.title('Median Blur')
plt.axis('off')
plt.show()
    
```

## Bilateral Filter

```python

# Apply Bilateral Filter
bilateral_filter = cv2.bilateralFilter(image_rgb, 9, 75, 75)

# Display the result
plt.imshow(bilateral_filter)
plt.title('Bilateral Filter')
plt.axis('off')
plt.show()
    
```

## Comparison of Results

```python

# Display all results side by side for comparison
fig, axs = plt.subplots(5, 1, figsize=(10, 20))

axs[0].imshow(image_rgb)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(average_blur)
axs[1].set_title('Average Blur')
axs[1].axis('off')

axs[2].imshow(gaussian_blur)
axs[2].set_title('Gaussian Blur')
axs[2].axis('off')

axs[3].imshow(median_blur)
axs[3].set_title('Median Blur')
axs[3].axis('off')

axs[4].imshow(bilateral_filter)
axs[4].set_title('Bilateral Filter')
axs[4].axis('off')

plt.tight_layout()
plt.show()
    
```
